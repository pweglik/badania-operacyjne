```python
import numpy as np
import matplotlib.pyplot as plt

for i, key in enumerate(keys):
    values = gs[key].unique()
    for val in values:
        fitness = gs[gs[key] == val]["fitness"]
        hist, bins = np.histogram(fitness, bins=30)
        plt.hist(fitness, bins=bins, label=f"{key}={val}", alpha=0.5, rwidth=1)

    plt.xlabel("funkcja dopasowania")
    plt.ylabel("liczba testÃ³w")
    plt.title(f"dopasowanie wzglÄ™dem {key}")
    plt.legend()
    plt.savefig(f"{i+1}.png")
    plt.show()
```
