```python
import pandas as pd

gs = pd.read_csv("gridsearch2.csv")
```

```python
gs.head()
```

```python
keys = [
    "survival_functions",
    "chance_create_line",
    "chance_cycle",
    "chance_erase_line",
    "chance_invert",
    "chance_merge",
    "chance_merge_specimen",
    "chance_rot_cycle",
    "chance_rot_right",
    "chance_split",
]
```

```python
import numpy as np
import matplotlib.pyplot as plt

for key in keys:
    values = gs[key].unique()
    for val in values:
        fitness = gs[gs[key] == val]["fitness"]
        hist, bins = np.histogram(fitness, bins=30)
        plt.hist(fitness, bins=bins, label=f"{key}={val}", alpha=0.5, rwidth=1)

    plt.xlabel("funkcja dopasowania")
    plt.ylabel("liczba testw")
    plt.title(f"dopasowanie wzgldem {key}")
    plt.legend()
    plt.show()
```
